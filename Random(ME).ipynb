# Uso de números aleatorios

```python
import random as rd
rd.random()
```

Ejercicio 1. Simular un volado.


```python
def volado():
  if rd.random() < 0.5:
    return "Cara"
  else:
    return "Cruz"
```

```python
for i in range(10):
  print(volado())
```

Ejercicio 2. Use la función anterior para estimar la probabilidad de obtener "cara"

```python
def aux():
  if volado() == "Cara":
    return 1
  else:
    return 0

n = 10000000
tsum = 0
for i in range(n):
  tsum += aux()
print(tsum/n)
```

Caminatas al azar

```python
import random as rd
import matplotlib.pyplot as plt


def generar_paso(p):
    """
    Decide si el paso es hacia arriba o abajo.
    p: probabilidad de avanzar (+1)
    """
    if rd.random() < p:
        return 1
    else:
        return -1


def simular_caminata(n_pasos, p):
    """
    Genera la lista de posiciones acumuladas.
    n_pasos: cantidad de movimientos
    p: probabilidad de avance
    """
    posicion_actual = 0
    trayectoria = [posicion_actual]

    for _ in range(n_pasos):
        paso = generar_paso(p)
        posicion_actual += paso
        trayectoria.append(posicion_actual)

    return trayectoria


n = 1000
prob_p = 0.6
datos_caminata = simular_caminata(n, prob_p)

plt.figure(figsize=(10, 5))
plt.plot(datos_caminata, label=f'Trayectoria con $p={prob_p}$')
plt.axhline(0, color='black', linestyle='--', alpha=0.3)


plt.title(r'Caminata Aleatoria: $X_n = \sum_{i=1}^{n} Z_i$ donde $P(Z_i=1)=p$')
plt.xlabel('Pasos ($n$)')
plt.ylabel('Posición ($X_n$)')
plt.legend()
plt.grid(True, alpha=0.2)
plt.show()
```
