## CAMINARA ALEATORIA 1D

Tenemos un punto que se puede mover a la derecha con probabilidad $p$ y a la izquierda con probabilidad $q$ tal que $$p+q=1$$

```python
import random as rd
import matplotlib.pyplot as plt

npasos = 100
p = 0.57
posicion = 0
trayectoria = [posicion]

def pasos(p):
  if rd.random() < p:
    return 1
  else:
    return -1


def caminata(npasos,p):
  posicion = 0
  trayectoria = [posicion]
  for i in range(npasos):
    posicion += pasos(p)
    trayectoria.append(posicion)
  return trayectoria

trayectoria = caminata(npasos,p)

plt.figure(figsize=(10, 5))
plt.plot(trayectoria, color='purple')
plt.axhline(0, color='blue', linestyle=':', alpha=0.6)
plt.title(f'Caminata al Azar ($p={p}$)')
plt.xlabel('Pasos')
plt.ylabel('Posición')
plt.show()
```

# Analisis
Sea $X_i$ una variable aleatoria que representa el paso dado en el instante $i$ Definimos sus probabilidades como:

$$X_i = \begin{cases} 1 & \text{si } \text{rd.random()} < p \\ -1 & \text{en otro caso} \end{cases}$$


# Simulación de Caminata Aleatoria
La posición final está dada por la sumatoria:
$$Z_n = \sum_{i=1}^{n} X_i$$

Donde la probabilidad de éxito es $p$ y la de fracaso es $q$:
$$p + q = 1$$

Se implementan las funciones:
1. `pasos(p)` $\rightarrow  X_i$
2. `caminata(n, p)` $\rightarrow Z_n$


# Preguntas a plantear

1.   ¿El punto regresa al origen?




```python
p = 0.5
caminata1 = caminata(100, p)

def retorno(trayectoria):
  trayecto=trayectoria
  for i in range(1,len(trayecto)):
    if trayecto[i] == 0:
      return i
  return None

def resultado(caminata):
  pasoreturn = retorno(caminata)
  if retorno(caminata):
    print(f"regreso en {pasoreturn}")
  else:
    print("no regreso")


resultado(caminata1)
```

Elaboramos una funcion que detecta si en primera instancia regresa al origen y en que paso regreso pero con $p=0.5$, ahora veamos cual es su comportamiento si $p>0.5$ y ver si conforme crece la probabilidad de $p$ no es más o menos probable que retorne a 0

Hacemos una funcion que nos calcule la probabilidad de retornar con la simulacion de la ley de los grandes números

```python
def prob_return(n, npasos, p):
    exitos = 0
    for _ in range(n):

        camino = caminata(npasos, p)
        if retorno(camino):
            exitos += 1

    return exitos / n
```

**Simulaciones** $$p=0.5, p=0.65 , p=0.75,p=0.85$$

```python
p1=0.5
caminata2 = caminata(100, p)
pasoreturn=retorno(caminata2)
resultado(caminata2)
prob_return(500, 100, p1)


```

```python
p2=0.65
caminata3 = caminata(100, p)
pasoreturn=retorno(caminata3)
resultado(caminata3)
prob_return(500, 100, p2)
```

```python
p3=0.75
caminata4 = caminata(100, p)
pasoreturn=retorno(caminata4)
resultado(caminata4)
prob_return(500, 100, p3)
```

```python
p4=0.85
caminata5 = caminata(100, p)
pasoreturn=retorno(caminata5)
resultado(caminata5)
prob_return(500, 100, p4)
```

Se observa que:
cada vez que la probabilidad de $p$ va creciendo la probabilidad de $retornar$ va disminullendo.

**Concluimos que:**

Esta caminata es una cadena de Markov pues el futuro depende solo del presente.
